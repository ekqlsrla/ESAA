### Step 5. Transform the Date column as a datetime type

```python
apple = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv',sep=',',parse_dates=[0])
```

### Step 6.  Set the date as the index

```python
apple.set_index('Date')
```

```python
apple.index = apple['Date']

apple
```

### Step 7.  Is there any duplicate dates?

```python
apple[apple.duplicated('Date')]

#No There isn't any duplicate dates
```

### Step 8.  Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

```python
apple.sort_values(by = 'Date',ascending = True).reset_index(drop=True)
```

```python
apple.iloc[::-1]
```

### Step 9. Get the last business day of each month

```python
apple['month'] = apple['Date'].dt.month
apple['Day'] = apple['Date'].dt.day

apple.groupby('month')['Day'].max()
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
apple.Date.max() - apple.Date.min()
```

### Step 11.  How many months in the data we have?

```python
apple.month.count()
```

### Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches

```python
plt.figure(figsize = (13.5,9))
plt.hist('Adj Close',data = apple)
plt.show()
```
